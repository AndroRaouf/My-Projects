```python
!pip install rapidfuzz --quiet

import pandas as pd, numpy as np, re
from rapidfuzz import process, fuzz
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid", font_scale=1.1)

# ÿ™ÿπÿ±ŸäŸÅ ÿßŸÑŸÅÿßŸäŸÑÿßÿ™
AMAZON_FILE = "amazon_smart_watches.csv"
NOON_FILE = "noon_smart_watches.csv"
MATCH_THRESHOLD = 70
BATCH_SIZE = 6

# ŸÇÿ±ÿßÿ°ÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
amazon = pd.read_csv(AMAZON_FILE)
noon = pd.read_csv(NOON_FILE)

print(f"üìä Amazon Smart Watches: {amazon.shape[0]} products")
print(f"üìä Noon Smart Watches: {noon.shape[0]} products")
print("\nüîç Amazon columns:", amazon.columns.tolist())
print("üîç Noon columns:", noon.columns.tolist())

def drop_junk_cols(df):
    """ÿ≠ÿ∞ŸÅ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿ∫Ÿäÿ± ÿßŸÑŸÖŸáŸÖÿ©"""
    mask = ~df.columns.str.contains("src|\(\d+\)|Unnamed|placeholder|image|carousel|link", regex=True, case=False)
    return df.loc[:, mask]

def find_column_like(df, patterns):
    """ÿßŸÑÿ®ÿßÿ≠ÿ´ ÿπŸÜ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿ®ŸÜÿßÿ° ÿπŸÑŸâ ÿ£ŸÜŸÖÿßÿ∑"""
    cols = df.columns.tolist()
    for p in patterns:
        for c in cols:
            if p.lower() in c.lower():
                return c
    text_cols = [c for c in cols if df[c].dtype == object]
    if text_cols:
        return text_cols[0]
    return None

def clean_numeric_series(s):
    """ÿ™ŸÜÿ∏ŸäŸÅ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ±ŸÇŸÖŸäÿ©"""
    if s is None or len(s) == 0:
        return pd.Series([np.nan] * len(s))
    s = s.astype(str).str.replace(",", "", regex=False).str.strip()
    s = s.str.replace("EGP", "", regex=False).str.strip()
    s = s.str.replace("out of 5 stars", "", regex=False).str.strip()
    extracted = s.str.extract(r"([0-9]+(?:\.[0-9]+)?)")[0]
    return pd.to_numeric(extracted, errors="coerce")

def normalize_product_name(s):
    """ÿ™Ÿàÿ≠ŸäÿØ ÿ™ŸÜÿ≥ŸäŸÇ ÿ£ÿ≥ŸÖÿßÿ° ÿßŸÑŸÖŸÜÿ™ÿ¨ÿßÿ™"""
    s = str(s).lower()
    # ÿ•ÿ≤ÿßŸÑÿ© ÿßŸÑŸÉŸÑŸÖÿßÿ™ ÿßŸÑÿ¥ÿßÿ¶ÿπÿ© ŸÅŸä ÿßŸÑÿ≥ÿßÿπÿßÿ™ ÿßŸÑÿ∞ŸÉŸäÿ©
    s = re.sub(r"\b(series|watch|smartwatch|smart watch|bluetooth|call|waterproof|ip68|fitness|tracker|android|ios)\b", " ", s)
    s = re.sub(r"\b(\d+\s*(inch|mm|mah|atm|hours|days|months|years))\b", " ", s)
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def extract_brand_from_name(product_name):
    """ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿ®ÿ±ÿßŸÜÿØ ŸÖŸÜ ÿßÿ≥ŸÖ ÿßŸÑŸÖŸÜÿ™ÿ¨"""
    brands = ['oraimo', 'xiaomi', 'redmi', 'huawei', 'samsung', 'apple', 'amazfit', 'cmf', 'infinix', 'honor', 'realme', 'adhomax', 'generic', 'joyroom', 'merv', 'nabi']
    product_lower = str(product_name).lower()
    for brand in brands:
        if brand in product_lower:
            return brand.title()
    return "Other"

# ÿ™ŸÜÿ∏ŸäŸÅ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
amazon = drop_junk_cols(amazon)
noon = drop_junk_cols(noon)

print(f"\nüîÑ After cleaning - Amazon: {amazon.shape}, Noon: {noon.shape}")

# ÿ™ÿ≠ÿØŸäÿØ ÿßŸÑÿ£ÿπŸÖÿØÿ©
a_product_col = "Product_Name"
a_price_col = "Price"
a_rating_col = "Rating"
a_reviews_col = "Reviews"

n_product_col = "Product_Name"
n_price_col = "Price"
n_old_col = "Old_Price"
n_rating_col = "Rating"
n_reviews_col = "Reviews"

# ÿ™ŸÜÿ∏ŸäŸÅ ÿ®ŸäÿßŸÜÿßÿ™ Amazon
amazon_clean = pd.DataFrame()
amazon_clean['Product_raw'] = amazon[a_product_col] if a_product_col in amazon.columns else amazon.iloc[:, 0]
amazon_clean['Price_raw'] = amazon[a_price_col] if a_price_col in amazon.columns else amazon.iloc[:, 6]
amazon_clean['Rating_raw'] = amazon[a_rating_col] if a_rating_col in amazon.columns else amazon.iloc[:, 4]
amazon_clean['Reviews_raw'] = amazon[a_reviews_col] if a_reviews_col in amazon.columns else amazon.iloc[:, 5]

# ÿ™ŸÜÿ∏ŸäŸÅ ÿ®ŸäÿßŸÜÿßÿ™ Noon
noon_clean = pd.DataFrame()
noon_clean['Product_raw'] = noon[n_product_col] if n_product_col in noon.columns else noon.iloc[:, 3]
noon_clean['Price_raw'] = noon[n_price_col] if n_price_col in noon.columns else noon.iloc[:, 6]
noon_clean['OldPrice_raw'] = noon[n_old_col] if n_old_col in noon.columns else (noon.iloc[:, 7] if len(noon.columns) > 7 else None)
noon_clean['Rating_raw'] = noon[n_rating_col] if n_rating_col in noon.columns else noon.iloc[:, 4]
noon_clean['Reviews_raw'] = noon[n_reviews_col] if n_reviews_col in noon.columns else noon.iloc[:, 5]

# ÿ™ÿ∑ÿ®Ÿäÿπ ÿ£ÿ≥ŸÖÿßÿ° ÿßŸÑŸÖŸÜÿ™ÿ¨ÿßÿ™
amazon_clean['Product_norm'] = amazon_clean['Product_raw'].fillna("").map(normalize_product_name)
noon_clean['Product_norm'] = noon_clean['Product_raw'].fillna("").map(normalize_product_name)

# ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿ®ÿ±ÿßŸÜÿØ
amazon_clean['Brand'] = amazon_clean['Product_raw'].map(extract_brand_from_name)
noon_clean['Brand'] = noon_clean['Product_raw'].map(extract_brand_from_name)

# ÿ™ŸÜÿ∏ŸäŸÅ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ±ŸÇŸÖŸäÿ©
amazon_clean['Price'] = clean_numeric_series(amazon_clean['Price_raw'])
amazon_clean['Rating'] = clean_numeric_series(amazon_clean['Rating_raw'])
amazon_clean['Reviews'] = clean_numeric_series(amazon_clean['Reviews_raw'])

noon_clean['Price'] = clean_numeric_series(noon_clean['Price_raw'])
noon_clean['Old_Price'] = clean_numeric_series(noon_clean['OldPrice_raw']) if noon_clean['OldPrice_raw'] is not None else pd.Series([np.nan] * len(noon_clean))
noon_clean['Rating'] = clean_numeric_series(noon_clean['Rating_raw'])
noon_clean['Reviews'] = clean_numeric_series(noon_clean['Reviews_raw'])

print(f"\n‚úÖ Amazon cleaned: {len(amazon_clean)} smart watches")
print(f"‚úÖ Noon cleaned: {len(noon_clean)} smart watches")

# ÿπÿ±ÿ∂ ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ®ÿ±ÿßŸÜÿØÿßÿ™
print(f"\nüè∑Ô∏è Top Brands Distribution:")
print("Amazon brands:", amazon_clean['Brand'].value_counts().head(5))
print("Noon brands:", noon_clean['Brand'].value_counts().head(5))

# ÿπŸÖŸÑ ÿßŸÑŸÖÿ∑ÿßÿ®ŸÇÿ© ÿ®ŸäŸÜ ÿßŸÑŸÖŸÜÿ™ÿ¨ÿßÿ™
noon_choices = noon_clean['Product_norm'].fillna("").tolist()
results = []

print(f"\nüîç Matching smart watches with threshold: {MATCH_THRESHOLD}%")

for idx, row in amazon_clean.iterrows():
    a_name = row['Product_norm']
    if not a_name.strip():
        continue

    match = process.extractOne(a_name, noon_choices, scorer=fuzz.token_set_ratio)
    if match is None:
        continue

    best_name, score, match_index = match
    if score < MATCH_THRESHOLD:
        continue

    n_row = noon_clean.iloc[match_index]
    a_price = row['Price'] if not pd.isna(row['Price']) else np.nan
    n_price = n_row['Price'] if not pd.isna(n_row['Price']) else np.nan

    # ÿ≠ÿ≥ÿßÿ® ŸÅÿ±ŸÇ ÿßŸÑÿ≥ÿπÿ± (Amazon - Noon)
    price_diff = (a_price - n_price) if (not np.isnan(a_price) and not np.isnan(n_price)) else np.nan

    # ÿ™ÿ≠ÿØŸäÿØ ÿßŸÑÿ£ŸÅÿ∂ŸÑ ÿ®ŸÜÿßÿ° ÿπŸÑŸâ ÿßŸÑÿ≥ÿπÿ±
    if np.isnan(price_diff):
        if not np.isnan(a_price) and np.isnan(n_price):
            best_opt = "Amazon"
        elif np.isnan(a_price) and not np.isnan(n_price):
            best_opt = "Noon"
        else:
            best_opt = "Unknown"
    else:
        best_opt = "Amazon" if a_price < n_price else "Noon"

    results.append({
        "Amazon_idx": idx,
        "Noon_idx": match_index,
        "Brand": row['Brand'],
        "Amazon_Product": row['Product_raw'],
        "Amazon_Price": a_price,
        "Amazon_Rating": row['Rating'],
        "Amazon_Reviews": row['Reviews'],
        "Noon_Product": n_row['Product_raw'],
        "Noon_Price": n_price,
        "Noon_Old_Price": n_row['Old_Price'],
        "Noon_Rating": n_row['Rating'],
        "Noon_Reviews": n_row['Reviews'],
        "Match_Score": score,
        "Price_Diff_Amazon_Minus_Noon": price_diff,
        "Best_Option": best_opt
    })

# ÿ•ŸÜÿ¥ÿßÿ° DataFrame ÿßŸÑŸÜŸáÿßÿ¶Ÿä
comparison = pd.DataFrame(results).sort_values("Match_Score", ascending=False).reset_index(drop=True)

# ÿ•ÿ∂ÿßŸÅÿ© ÿπŸÖŸàÿØ ŸäŸàÿ∂ÿ≠ ŸÅÿ±ŸÇ ÿßŸÑÿ≥ÿπÿ± ÿ®ÿ¥ŸÉŸÑ ÿ£Ÿàÿ∂ÿ≠
comparison['Price_Diff_EGP'] = comparison['Price_Diff_Amazon_Minus_Noon'].apply(
    lambda x: f"Amazon +{abs(x):.0f} EGP" if x > 0 else f"Noon +{abs(x):.0f} EGP" if x < 0 else "Same Price" if x == 0 else "N/A"
)

# ÿ≠ŸÅÿ∏ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
comparison.to_csv("Amazon_vs_Noon_SmartWatches_comparison.csv", index=False)
comparison.to_excel("Amazon_vs_Noon_SmartWatches_comparison.xlsx", index=False)

print(f"\n‚úÖ Done. Total matches found: {len(comparison)}")

if len(comparison) > 0:
    print(f"\nüìä All matched Smart Watches ({len(comparison)} products):")

    # ÿπÿ±ÿ∂ ÿ¨ÿØŸàŸÑ ŸÖÿ®ÿ≥ÿ∑
    display_df = comparison[[
        'Brand',
        'Amazon_Product',
        'Noon_Product',
        'Match_Score',
        'Amazon_Price',
        'Noon_Price',
        'Price_Diff_Amazon_Minus_Noon',
        'Price_Diff_EGP',
        'Best_Option'
    ]].copy()

    # ÿ™ŸÇÿµŸäÿ± ÿ£ÿ≥ŸÖÿßÿ° ÿßŸÑŸÖŸÜÿ™ÿ¨ÿßÿ™ ŸÑŸÑÿπÿ±ÿ∂
    display_df['Amazon_Product'] = display_df['Amazon_Product'].str.slice(0, 50) + "..."
    display_df['Noon_Product'] = display_df['Noon_Product'].str.slice(0, 50) + "..."

    display(display_df.head(20))

    # ÿØÿßŸÑÿ© ŸÑÿπÿ±ÿ∂ ÿßŸÑÿ±ÿ≥ŸàŸÖ ÿßŸÑÿ®ŸäÿßŸÜŸäÿ©
    def plot_batches(df, batch_size=6):
        names = df["Amazon_Product"].astype(str).tolist()
        a_prices = df["Amazon_Price"].tolist()
        n_prices = df["Noon_Price"].tolist()
        price_diffs = df["Price_Diff_Amazon_Minus_Noon"].tolist()
        brands = df["Brand"].tolist()

        for i in range(0, len(names), batch_size):
            batch_names = names[i:i+batch_size]
            batch_a = a_prices[i:i+batch_size]
            batch_n = n_prices[i:i+batch_size]
            batch_diffs = price_diffs[i:i+batch_size]
            batch_brands = brands[i:i+batch_size]

            y = np.arange(len(batch_names))
            height = 0.35

            plt.figure(figsize=(16, max(6, len(batch_names)*1.0)))

            # ÿ™ŸÑŸàŸäŸÜ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿ≠ÿ≥ÿ® ÿßŸÑÿ®ÿ±ÿßŸÜÿØ
            colors_a = ['#1f77b4' if diff <= 0 else '#ff6b6b' for diff in batch_diffs]
            colors_n = ['#ff7f0e' if diff >= 0 else '#4ecdc4' for diff in batch_diffs]

            bars_a = plt.barh(y - height/2, batch_a, height=height, label="Amazon", color=colors_a, alpha=0.7)
            bars_n = plt.barh(y + height/2, batch_n, height=height, label="Noon", color=colors_n, alpha=0.7)

            # ÿ•ÿ∂ÿßŸÅÿ© ÿßŸÑÿ®ÿ±ÿßŸÜÿØ ŸÅŸä ÿßÿ≥ŸÖ ÿßŸÑŸÖŸÜÿ™ÿ¨
            y_labels = [f"{brand}\n{name[:45]}..." if len(name) > 45 else f"{brand}\n{name}"
                       for brand, name in zip(batch_brands, batch_names)]

            plt.yticks(y, y_labels, fontsize=8)
            plt.xlabel("Price (EGP)")
            plt.title(f"Amazon vs Noon Smart Watches Price Comparison - Batch {i//batch_size + 1}\n(Items {i+1}-{i+len(batch_names)})")

            # ÿ•ÿ∂ÿßŸÅÿ© ÿßŸÑŸÇŸäŸÖ ÿπŸÑŸâ ÿßŸÑÿ£ÿπŸÖÿØÿ©
            for yi, av, nv, diff in zip(y, batch_a, batch_n, batch_diffs):
                if not pd.isna(av):
                    plt.text(av, yi - height/2, f" {av:.0f}", va="center", fontsize=7, fontweight='bold')
                if not pd.isna(nv):
                    plt.text(nv, yi + height/2, f" {nv:.0f}", va="center", fontsize=7, fontweight='bold')

                # ÿ•ÿ∂ÿßŸÅÿ© ŸÅÿ±ŸÇ ÿßŸÑÿ≥ÿπÿ± ŸÅŸä ÿßŸÑŸÖŸÜÿ™ÿµŸÅ
                if not pd.isna(diff):
                    mid_x = min(av, nv) + abs(diff)/2
                    color = "green" if diff < 0 else "red" if diff > 0 else "blue"
                    diff_text = f"Œî={diff:.0f}"
                    plt.text(mid_x, yi, diff_text, va="center", ha="center",
                            fontsize=8, fontweight='bold', color=color,
                            bbox=dict(boxstyle="round,pad=0.2", facecolor="white", alpha=0.8))

            plt.legend()
            plt.tight_layout()
            plt.show()

    # ÿπÿ±ÿ∂ ÿßŸÑÿ±ÿ≥ŸàŸÖ ÿßŸÑÿ®ŸäÿßŸÜŸäÿ©
    plot_df = comparison.sort_values("Match_Score", ascending=False).reset_index(drop=True)
    plot_batches(plot_df, batch_size=BATCH_SIZE)

    # üÜï ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ŸÖÿ™ŸÇÿØŸÖÿ© Ÿàÿ•ÿ∂ÿßŸÅÿßÿ™ ŸÉÿ±Ÿäÿ™ŸäŸÅ
    print("\n" + "="*50)
    print("üéØ SMART WATCHES COMPARISON INSIGHTS")
    print("="*50)

    print(f"\nüìà Basic Statistics:")
    print(f"Total matches: {len(comparison)}")
    print(f"Average match score: {comparison['Match_Score'].mean():.1f}%")

    amazon_cheaper = len(comparison[comparison['Best_Option'] == 'Amazon'])
    noon_cheaper = len(comparison[comparison['Best_Option'] == 'Noon'])
    same_price = len(comparison[(comparison['Price_Diff_Amazon_Minus_Noon'] == 0) & (~comparison['Price_Diff_Amazon_Minus_Noon'].isna())])

    print(f"üì¶ Amazon cheaper in: {amazon_cheaper} smart watches")
    print(f"üõí Noon cheaper in: {noon_cheaper} smart watches")
    print(f"‚öñÔ∏è Same price in: {same_price} smart watches")

    # üÜï ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿ®ÿ±ÿßŸÜÿØÿßÿ™
    print(f"\nüè∑Ô∏è Brand Analysis:")
    brand_stats = comparison.groupby('Brand').agg({
        'Price_Diff_Amazon_Minus_Noon': 'mean',
        'Match_Score': 'mean',
        'Amazon_Price': 'mean',
        'Best_Option': lambda x: (x == 'Amazon').sum()
    }).round(2)

    brand_stats.columns = ['Avg_Price_Diff', 'Avg_Match_Score', 'Avg_Price', 'Amazon_Wins']
    brand_stats['Noon_Wins'] = len(comparison) - brand_stats['Amazon_Wins']
    print(brand_stats.sort_values('Avg_Price_Diff', ascending=True))

    # üÜï ÿ£ŸÅÿ∂ŸÑ ÿßŸÑÿµŸÅŸÇÿßÿ™ ÿ≠ÿ≥ÿ® ÿßŸÑŸÅÿ¶ÿ© ÿßŸÑÿ≥ÿπÿ±Ÿäÿ©
    print(f"\nüíé Best Deals by Price Category:")

    # ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ≥ÿßÿπÿßÿ™ ÿßŸÑÿ∞ŸÉŸäÿ© ŸÑÿ¥ÿ±ÿßÿ¶ÿ≠ ÿ≥ÿπÿ±Ÿäÿ©
    comparison['Price_Category'] = pd.cut(comparison['Amazon_Price'],
                                        bins=[0, 1000, 3000, 8000, float('inf')],
                                        labels=['Budget (<1K)', 'Mid-Range (1K-3K)', 'Premium (3K-8K)', 'Luxury (>8K)'])

    best_deals_by_category = []
    for category in comparison['Price_Category'].unique():
        if pd.isna(category):
            continue
        category_data = comparison[comparison['Price_Category'] == category]
        if len(category_data) > 0:
            best_amazon = category_data[category_data['Price_Diff_Amazon_Minus_Noon'] == category_data['Price_Diff_Amazon_Minus_Noon'].min()]
            best_noon = category_data[category_data['Price_Diff_Amazon_Minus_Noon'] == category_data['Price_Diff_Amazon_Minus_Noon'].max()]

            if len(best_amazon) > 0:
                best_deals_by_category.append({
                    'Category': category,
                    'Store': 'Amazon',
                    'Product': best_amazon.iloc[0]['Amazon_Product'][:40] + "...",
                    'Price_Diff': best_amazon.iloc[0]['Price_Diff_Amazon_Minus_Noon'],
                    'Amazon_Price': best_amazon.iloc[0]['Amazon_Price'],
                    'Noon_Price': best_amazon.iloc[0]['Noon_Price']
                })

            if len(best_noon) > 0:
                best_deals_by_category.append({
                    'Category': category,
                    'Store': 'Noon',
                    'Product': best_noon.iloc[0]['Amazon_Product'][:40] + "...",
                    'Price_Diff': best_noon.iloc[0]['Price_Diff_Amazon_Minus_Noon'],
                    'Amazon_Price': best_noon.iloc[0]['Amazon_Price'],
                    'Noon_Price': best_noon.iloc[0]['Noon_Price']
                })

    deals_df = pd.DataFrame(best_deals_by_category)
    if not deals_df.empty:
        display(deals_df.sort_values(['Category', 'Price_Diff']))

    # üÜï ÿ™ŸàÿµŸäÿßÿ™ ÿßŸÑÿ¥ÿ±ÿßÿ°
    print(f"\nü§î Smart Buying Recommendations:")

    if amazon_cheaper > noon_cheaper:
        print("üí° Overall, Amazon offers better prices for smart watches")
    elif noon_cheaper > amazon_cheaper:
        print("üí° Overall, Noon offers better prices for smart watches")
    else:
        print("üí° Prices are fairly balanced between Amazon and Noon")

    # ÿ™ŸàÿµŸäÿßÿ™ ÿ≠ÿ≥ÿ® ÿßŸÑÿ®ÿ±ÿßŸÜÿØ
    print(f"\nüèÜ Top Brand Recommendations:")
    for brand in comparison['Brand'].value_counts().head(3).index:
        brand_data = comparison[comparison['Brand'] == brand]
        amazon_wins = len(brand_data[brand_data['Best_Option'] == 'Amazon'])
        noon_wins = len(brand_data[brand_data['Best_Option'] == 'Noon'])

        if amazon_wins > noon_wins:
            print(f"  ‚Ä¢ {brand}: Better on Amazon ({amazon_wins} vs {noon_wins})")
        elif noon_wins > amazon_wins:
            print(f"  ‚Ä¢ {brand}: Better on Noon ({noon_wins} vs {amazon_wins})")
        else:
            print(f"  ‚Ä¢ {brand}: Balanced between both platforms")

    # üÜï ŸÖÿÆÿ∑ÿ∑ ÿ•ÿ∂ÿßŸÅŸä: ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ£ÿ≥ÿπÿßÿ± ÿ≠ÿ≥ÿ® ÿßŸÑÿ®ÿ±ÿßŸÜÿØ
    plt.figure(figsize=(12, 8))

    # ÿ™ÿ¨ŸÖŸäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÑŸÑÿ±ÿ≥ŸÖ
    plot_data = []
    for idx, row in comparison.iterrows():
        if not pd.isna(row['Amazon_Price']):
            plot_data.append({'Brand': row['Brand'], 'Platform': 'Amazon', 'Price': row['Amazon_Price']})
        if not pd.isna(row['Noon_Price']):
            plot_data.append({'Brand': row['Brand'], 'Platform': 'Noon', 'Price': row['Noon_Price']})

    if plot_data:
        plot_df = pd.DataFrame(plot_data)

        plt.subplot(1, 2, 1)
        sns.boxplot(data=plot_df, x='Brand', y='Price', hue='Platform')
        plt.title('Price Distribution by Brand and Platform')
        plt.xticks(rotation=45)
        plt.ylabel('Price (EGP)')

        plt.subplot(1, 2, 2)
        price_comparison = comparison[['Brand', 'Amazon_Price', 'Noon_Price']].melt(id_vars=['Brand'],
                                                                                   value_vars=['Amazon_Price', 'Noon_Price'],
                                                                                   var_name='Platform', value_name='Price')
        sns.violinplot(data=price_comparison.dropna(), x='Platform', y='Price')
        plt.title('Overall Price Distribution')
        plt.ylabel('Price (EGP)')

        plt.tight_layout()
        plt.show()

else:
    print("‚ùå No matches found. Try lowering the MATCH_THRESHOLD value.")

print("\nüíæ Saved files:")
print("- Amazon_vs_Noon_SmartWatches_comparison.csv")
print("- Amazon_vs_Noon_SmartWatches_comparison.xlsx")

print(f"\nüéâ Comparison completed successfully! Found {len(comparison)} matching smart watches.")
```
